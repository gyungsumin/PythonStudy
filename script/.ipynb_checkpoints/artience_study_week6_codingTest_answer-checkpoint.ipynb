##### Quiz 3. Ï†ÄÏàòÏö©Îüâ Í≥ÑÏÇ∞ÌïòÍ∏∞ (Hard) - Î™ª Ìë∏ÎäîÍ≤å Ï†ïÏÉÅ


- ÎÜíÏù¥ Í∞íÎì§ÏùÑ ÏûÖÎ†•Î∞õÏïÑ, ÎπÑÍ∞Ä Ïò¨ Îïå ÏµúÎåÄ Ï†ÄÏàòÏö©Îüâ(Î∂ÄÌîº)ÏùÑ Í≥ÑÏÇ∞ÌïòÎäî Ìï®Ïàò ÎßåÎì§Í∏∞
- ÎÑàÎπÑ(width)ÏôÄ Í∏∏Ïù¥(length)Îäî Í≥†Ï†ï
- ÏïÑÎûò ÏòàÏãúÏóêÏÑúÏùò Ï†ïÎãµÏùÄ 6.0


- input example
-- height_list = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1, 0]
-- width = 1
-- length = 1


- output
-- dam_load_volume() that returns single numeric

```python
# Quiz 3 answer

def dam_load_volume(height_list, width, length):
    max_height = max(height_list)
    
    # Í∑πÎã®Ïùò ÎÜíÏù¥Í∞Ä 0Ïùº Í≤ΩÏö∞, damÏù¥ ÏïÑÎãå ÌèâÏßÄÏôÄ ÎßàÏ∞¨Í∞ÄÏßÄ
    if height_list[0] == 0:
        del height_list[0]
    if height_list[-1] == 0:
        del height_list[-1]

        
    # Î∂ÄÌîº
    volume = 0
    
    # ÏµúÎåÄ ÎÜíÏù¥ ~ ÏµúÏ†Ä ÎÜíÏù¥ loop
    for layer in range(max_height, 0, -1):
        dam_idx = [i for i, x in enumerate(height_list) if x >= layer]  # ÌòÑÏû¨ ÎÜíÏù¥Í∞Ä damÏù∏ ÏßÄÏ†ê list
        area_idx = [i for i, x in enumerate(height_list) if x < layer]  # ÌòÑÏû¨ ÎÜíÏù¥Í∞Ä areaÏù∏ ÏßÄÏ†ê list

        if len(dam_idx) == 1:  # ÌòÑ ÎÜíÏù¥Ïóê damÏù¥ ÌïòÎÇòÏùº Í≤ΩÏö∞ -> Î¨¥Ï°∞Í±¥ ÏµúÍ≥†Ï†ê. (ÏµúÍ≥†Ï†êÏù¥ Ïñ∏Ï†úÎÇò ÌïòÎÇòÏù∏Í±¥ ÏïÑÎãò)
            continue  # ÏµúÍ≥†Ï†ê damÏù¥ ÌïòÎÇòÎßå ÏûàÏùÑ Í≤ΩÏö∞, Í∑∏ ÎÜíÏù¥ÏóêÏÑúÎäî Ï†ÅÏû¨ Î∂àÍ∞Ä
        else:
            for idx in range(len(dam_idx)-1):
                left_dam = dam_idx[idx]
                right_dam = dam_idx[idx + 1]

                possible_area = range(left_dam + 1, right_dam)

                volume += ((len(possible_area) * width) * length * 1)  # Í∏∏Ïù¥ x ÎÑàÎπÑ x ÎÜíÏù¥(1 layer ÎÜíÏù¥Îäî Ïñ∏Ï†úÎÇò 1)
                
    return volume
```

```python
height_list = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1, 0]
width = 1
length = 1
dam_load_volume(height_list=height_list, width=width, length=length)
```
