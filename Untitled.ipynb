```python
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium_stealth import stealth
import pyautogui

import time

from app.main.module.api.cm360 import connection

from app.main.config.env_config import CM360_REQUEST_XLSX_FILE
from app.main.config.env_config import CHROMEDRIVER
```

```python



# 다운받은 파일 전처리, api에서 이름정보 불러와서 파일명 바꿔주는 용도로 사용하기

def run(account_id, profile_id, advertiser_id):
    options = __options__(account_id=account_id, profile_id=profile_id, advertiser_id=advertiser_id)

    service = connection.cm360_connection()
    campaign_list = service.campaigns().list(profileId=profile_id).execute()["campaigns"]
    campaign_list = [i for i in campaign_list if i["archived"] is False]

    campaign_list_survived = []
    for idx in options.index:
        elements = options.loc[idx][["project", "country", "year", "quarter", "client", "cycle", "product"]]
        elements = set([str(i) for i in elements])
        elements.remove("IGNORE")

        temp = campaign_list.copy()
        for element in elements:
            temp = [i for i in temp if element in i["name"]]

        for i in temp:
            campaign_list_survived.append(i)

    print(type(campaign_list_survived))
    print(len(campaign_list_survived))
    print(campaign_list_survived)

    __robot_connect__(campaign_list=campaign_list_survived)


def __options__(account_id, profile_id, advertiser_id,
                file_path=CM360_REQUEST_XLSX_FILE, sheet_name="export_campaign"):

    option = pd.read_excel(file_path, sheet_name=sheet_name)

    option_account_id = str(int(option["cm account id"].dropna().iloc[0]))
    option_profile_id = str(int(option["cm profile id"].dropna().iloc[0]))
    option_advertiser_id = str(int(option["cm advertiser id"].dropna().iloc[0]))

    opion_export_segment = option[["project", "country", "year", "quarter", "client", "cycle", "product"]]

    # verify identifications
    verify = (account_id == option_account_id) & (profile_id == option_profile_id) &\
             (advertiser_id == option_advertiser_id)
    if verify is False:
        raise ValueError('ID NOT MATCHING. CHECK REQUEST OPTION: TEST OR ACTUAL Profile.')
    else:
        pass

    return opion_export_segment




def __robot_connect__(campaign_list):
    options = webdriver.ChromeOptions()
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    options.add_argument('--disable-blink-features=AutomationControlled')
    driver = webdriver.Chrome(CHROMEDRIVER, options=options)
    stealth(driver,
            languages=["en-US", "en"],
            vendor="Google Inc.",
            platform="Win32",
            webgl_vendor="Intel Inc.",
            renderer="Intel Iris OpenGL Engine",
            fix_hairline=True,
            )
    url = "https://campaignmanager.google.com/trafficking/#/accounts/{}/campaigns/{}/explorer?statuses=0;2"
    # url = "https://campaignmanager.google.com"
    for campaign in campaign_list[:1]:
        account_id = campaign["accountId"]
        campaign_id = campaign["id"]

        current_url = url.format(account_id, campaign_id)

        driver.get(current_url)

        driver.implicitly_wait(10)

        pyautogui.typewrite('artgsmin', interval=0.1)
        pyautogui.press('enter')
        time.sleep(3)

        pyautogui.typewrite('alsrudtn12!@', interval=0.2)
        pyautogui.press('enter')
        time.sleep(20)

        # profile_button = driver.find_elements(By.NAME, """account-menu""")
        # print(profile_button)
        # print(len(profile_button))al
        # [0].click()
        # driver.click()
        cnt = 0
        while driver.current_url != current_url:
            driver.get(current_url)
            cnt += 1
            print(cnt)

            if cnt >= 10:
                raise RuntimeError("URL Access Error : {}".format(current_url))

        test = driver.find_elements(By.CLASS_NAME, "ddm-menubutton ng-scope ddm-managed-panel")
        print(test)
        test[0].click()
        # print(driver.current_url)
        # driver.get(current_url)
        # print(driver.current_url)
        time.sleep(10)

        time.sleep(10000)


print(1)
```
